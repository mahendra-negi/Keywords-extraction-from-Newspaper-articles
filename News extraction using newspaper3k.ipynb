```python
import pandas as pd
import numpy as np
import newspaper
```

```python
news_Site=newspaper.build('https://indianexpress.com/')
```

```python
for article in news_Site.articles:
    print(article.url)
```

#### Extracting Source categories

```python
for category in news_Site.category_urls():
    print(category)
```

#### Extracting Source feeds

```python
for feed_url in news_Site.feed_urls():
    print(feed_url)
```

#### Newspaper3k also has some NLP features

Newspaper3k uses nltk (an another Python library) which does Newspaper3k's NLP work. nltk library is automatically installed when we install the Newspaper3k library. However, nltk needs to use a sub-package called "punkt"

```python
import nltk
nltk.download('punkt')
```
